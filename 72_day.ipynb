```python
import matplotlib.pyplot as plt
import pandas as pd
pf = pd.read_csv('QueryResults.csv', names=['DATE', 'TAG', 'POSTS'], header=0)
ali = pf.pivot(index='DATE', columns='TAG', values='POSTS')
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('SANALAR', fontsize=14)
plt.ylabel('Post sonlari', fontsize=14)

# for column in ali.columns:
#     plt.plot(ali.index, ali[column], ali.java)

# for column in ali.columns:
#     plt.plot(ali.index, ali[column], linewidth=3, label=ali[column].name)
#     plt.legend(fontsize=16) 

rol_ali = ali.rolling(window=6).mean()
for column in rol_ali.columns:
    plt.plot(rol_ali.index, rol_ali[column], linewidth=3, label=rol_ali[column].name)
plt.legend(fontsize=16)
```
